```python
import pandas as pd
```

```python
df = pd.read_csv('bios.csv')
print(df.shape)
df.head()
```

```python
df['Title(s)'].isnull().sum()
```

### Drop Columns That Are Not Useful

```python
df_useful_columns = df.drop(columns = ['Nick/petnames','Name order','Original name','Title(s)'])
df_useful_columns.head()
```

```python
df_useful_columns.drop(columns = 'Other names', inplace=True)
df_useful_columns.head()
```

```python
df_useful_columns.isnull().sum()
```

### Clean Syntax in 'Full name' and 'Used name'

```python
df_useful_columns['Used name'] = df_useful_columns['Used name'].str.replace('•', ' ', regex=False).str.strip()
df_useful_columns.head()
```

```python
df_useful_columns['Full name'] = df_useful_columns['Full name'].str.replace('•', ' ', regex=False).str.strip()
df_useful_columns['Full name']
```

```python
# Strip all characters other than letters from full name
df_useful_columns['Full name'] = df_useful_columns['Full name'].str.replace(r'"[^"]+"', '', regex=True).str.replace(r'\s+', ' ', regex=True).str.strip()
df_useful_columns.head()
```

```python
# Strip all characters other than letters from used name
df_useful_columns['Used name'] = df_useful_columns['Used name'].str.replace(r'"[^"]+"', '', regex=True).str.replace(r'\s+', ' ', regex=True).str.strip()
```

```python
df_useful_columns.head()
```

### Parse 'Born' Column into 'Birth Date', and 'Birth Place'

```python
# Handle edge cases in 'Born' column that don't follow typcial format date/place
df_born_edgecase = df_useful_columns[~df_useful_columns['Born'].str.contains(' in ', na=False) & df_useful_columns['Born'].notna()]
df_born_edgecase.sample(25)
```

```python
# regex that matches exactly four digits in a row indicating a year
df_useful_columns['Birth Date'] = df['Born'].str.extract(r'(\d{4})')

# Only extract if " in " is present, and take everything after in
df_useful_columns['Birth Place'] = df_useful_columns['Born'].str.extract(r'\bin\s+(.+)$')
df_useful_columns.head()
```

```python
# COMPLETED: Handle edge cases in 'Born' column that don't follow typcial format date/place
df_useful_columns[~df_useful_columns['Born'].str.contains(' in ', na=False) & df_useful_columns['Born'].notna()]
```

```python
# Drop the old 'Born' Column
df_useful_columns.drop(columns='Born', inplace=True)
df_useful_columns.head()
```

### Transform 'Died' Column to Show Only Date of Death

```python
df_useful_columns['Death Date'] = df['Died'].str.extract(r'(\d{4})')
df_useful_columns.head()
```

```python
# Drop the old 'Died' column
df_useful_columns.drop(columns = 'Died', inplace=True)
df_useful_columns.head()
```

### Transform 'Birth Date' and 'Death Date' years to numbers

```python
# Convert years to numbers
df_useful_columns['Birth Date'] = pd.to_numeric(df_useful_columns['Birth Date'], errors='coerce').astype('Int64')
df_useful_columns['Death Date'] = pd.to_numeric(df_useful_columns['Death Date'], errors='coerce').astype('Int64')
```

```python
df_useful_columns.dtypes
```

```python
df_useful_columns.head()
```

### Parse 'Measurements' Column into Separate 'Height' and 'Weight' Columns

```python
#Handle measurements that dont follow stardnard cm/kg format
df_useful_columns[~df_useful_columns['Measurements'].str.contains('/', na=False) & df_useful_columns['Measurements'].notna()]
```

```python
# Extract height only if followed by 'cm'
df_useful_columns['Height_cm'] = df_useful_columns['Measurements'].str.extract(r'(\d+)\s*cm')

# Extract weight only if followed by 'kg'
df_useful_columns['Weight_kg'] = df_useful_columns['Measurements'].str.extract(r'(\d+)\s*kg')

# Convert to numeric
df_useful_columns['Height_cm'] = pd.to_numeric(df_useful_columns['Height_cm'], errors='coerce')
df_useful_columns['Weight_kg'] = pd.to_numeric(df_useful_columns['Weight_kg'], errors='coerce')

df_useful_columns.head()
```

```python
# COMPLETED: Handle measurements that dont follow stardnard cm/kg format
df_useful_columns[~df_useful_columns['Measurements'].str.contains('/', na=False) & df_useful_columns['Measurements'].notna()]
```

```python
# Drop the old 'Measurements' column
df_useful_columns.drop(columns = 'Measurements', inplace =True)
df_useful_columns.head()
```

### Parse 'Birth Place' into City, Region, Country

```python
# Define regex pattern
pattern = r'^[^,]+,\s[^()]+ \([A-Z]{3}\)$'

# Create a new column that shows if the pattern matches
df_useful_columns['valid_format'] = df_useful_columns['Birth Place'].fillna('').str.match(pattern)
df_useful_columns.head(10)
```

```python
# View rows that do NOT match the pattern
invalid_rows = df_useful_columns[~df_useful_columns['valid_format']]
df_invalid_rows = df_useful_columns[df_useful_columns['valid_format'] == False]
print(df_invalid_rows.shape)
df_invalid_rows.head()
# We can leave rows with NaN for Birth Place
```

```python
# Regex expression to separate city, region, country
pattern = r'^([^,]+), ([^()]+) \(([A-Z]{3})\)$'

# Use str.extract to split into new columns
df_useful_columns[['Birth City', 'Birth Region', 'Birth Country']] = df_useful_columns['Birth Place'].str.extract(pattern)
df_useful_columns.head()
```

```python
df_useful_columns.isnull().sum()
```

### Drop Useless Columns

```python
# Drop 'Nationality' column due to so many NaN
# Drop old 'Birth Place' column and 'valid_format' column
df_useful_columns.drop(columns = ['Nationality', 'Birth Place', 'valid_format'], inplace=True)
```

```python
df_useful_columns.rename(columns={'athlete_id': 'Athlete ID', 'Height_cm': 'Height cm', 'Weight_kg': 'Weight kg'}, inplace=True)
df_useful_columns.head()
```

```python
df_useful_columns['Roles'] = df_useful_columns['Roles'].str.replace(r'"[^"]+"', '', regex=True).str.replace(r'\s+', ' ', regex=True).str.strip()
df_useful_columns['Roles'] = df_useful_columns['Roles'].str.replace('•', ' ', regex=False).str.strip()
df_useful_columns.head()
```

### Number of medals won in general grouped by NOC (weighted medal point score)

```python
# Create a subset df with NOC and Medals and filter out rows without medals
df_medals = df[['NOC', 'Medal']].dropna(subset=["Medal"])
print(df_medals.shape)
df_medals.head(10)
```

```python
# Medal types
df_medals.Medal.unique()
```

```python
# Point-based medal system
medal_points = {"Gold": 3, "Silver": 2, "Bronze": 1}

NOC_medals_points = (df_medals.assign(Points=lambda x: x["Medal"].map(medal_points)).groupby("NOC")["Points"].sum().sort_values(ascending=False).reset_index())
NOC_medals_points.head(10)
```

### Visualize the performance (most amount of points/medals) of the top 5 NOC's from the year 2000 to now

```python
# Create a subset df with Year, NOC, and Medal
df_points = df[['Year', 'NOC', 'Medal']].dropna(subset=["Medal"])
df_points.head()
```

```python
# Drop rows with Year<2000
df_points = df_points[df_points['Year'] >= 2000]
print(df_points.shape)
df_points.head()
```

```python
# Assign a points column
medal_points = {"Gold": 3, "Silver": 2, "Bronze": 1}

df_points['Points'] = df_points['Medal'].map(medal_points)
df_points.head()
```

```python
# Group by year and return the avg points
df_points = df_points.groupby(['Year', 'NOC'])['Points'].sum().reset_index(name="Points Scored")
df_points.head()
```

```python
# Order by ASC years and DESC points
df_points = df_points.sort_values(['Year', 'Points Scored'], ascending=[True, False])
```

```python
df_points.head()
```

```python
# Only display these results for the top 5 NOC's
NOC_medals_points.head(5)
```

```python
df_points_top5 = df_points[df_points['NOC'].isin(['USA', 'URS', 'GER', 'GBR', 'FRA'])].reset_index(drop=True)
df_points_top5.head(8)
```

```python
# So, URS has not placed since year 2000
```

```python
# Pivot so each team gets its own column
df_points_top5_pivot = df_points_top5.pivot(index='Year', columns='NOC', values='Points Scored')
```

```python
df_points_top5_pivot.head()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 5))
for NOC in df_points_top5_pivot.columns:
    plt.plot(df_points_top5_pivot.index, df_points_top5_pivot[NOC], marker='o', label=NOC)

plt.xlabel('Year')
plt.ylabel('Points Scored')
plt.title('NOC Performance Since 2000 (Medallion Points)')
plt.legend(title='NOC')
# Show every year as an x tick, increment by 2
plt.xticks(range(df_points_top5_pivot.index.min(), df_points_top5_pivot.index.max() + 1, 2), rotation=45)

plt.grid(True)
plt.show()
```

### For each NOC, which athlete has the most amount of 1st places?

```python
# Filter the 'results' table by only 1st places (already calculated earlier)
df_winners.head()
```

```python
athlete_winners_per_NOC = df_winners[['NOC', 'Athlete ID']]
athlete_winners_per_NOC.head()
```

```python
# Count the wins per athlete id and add column
athlete_winners_per_NOC['Wins per Athlete'] = athlete_winners_per_NOC.groupby(["NOC", "Athlete ID"])["Athlete ID"].transform("count")
athlete_winners_per_NOC.head()
```

```python
# Drop duplicate rows
athlete_winners_per_NOC = athlete_winners_per_NOC.drop_duplicates().reset_index(drop=True)
athlete_winners_per_NOC
```

```python
# Find max wins per NOC even if athletes are tied for max wins
max_wins = athlete_winners_per_NOC.groupby("NOC")["Wins per Athlete"].transform("max")

# Filter to only athletes with that max
top_athlete_winners = athlete_winners_per_NOC[athlete_winners_per_NOC["Wins per Athlete"] == max_wins]
```

```python
# Sort by atheletes witht the most wins per NOC
top_athlete_winners = top_athlete_winners.sort_values('Wins per Athlete', ascending=False)
top_athlete_winners.head(10)
```

```python
# Merge this dataframe with the bios dataframe to get the names of each athlete
# Bios dataframe
df_useful_columns.head()
```

```python
# Merge the two dataframes to get the names of each athlete
# Merge on 'Athlete ID' and keep all columns from top_athlete_winners df
merged_df = top_athlete_winners.merge(df_useful_columns[["Athlete ID", "Used name"]], on="Athlete ID", how="left")
print(merged_df.shape)
merged_df.head(15)
```

### Visualize the top 13 athletes performance and color code them by their NOC

```python
import matplotlib.pyplot as plt

# Take top 13 athletes by wins
top13 = merged_df.head(13)

# Assign colors to each NOC
NOCs = top13["NOC"].unique()
colors = plt.cm.tab20.colors  # a palette with 20 colors
team_color_map = {NOC: colors[i % len(colors)] for i, NOC in enumerate(NOCs)}

bar_colors = top13["NOC"].map(team_color_map)

# Create bar chart
plt.figure(figsize=(12,6))
plt.bar(top13["Used name"], top13["Wins per Athlete"], color=bar_colors)
plt.ylabel("Number of Wins")
plt.xlabel("Athlete Name")
plt.title("Top 13 Athletes by Wins")

# Rotate x-axis labels if needed
plt.xticks(rotation=45, ha='right')

# Create legend for teams
handles = [plt.Rectangle((0,0),1,1,color=team_color_map[n]) for n in NOCs]
plt.legend(handles, NOCs, title="NOC", bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()
```
